```python
null_values_coloumns = ['famsize','Fedu','traveltime','higher','freetime','G2','Feature_1','Feature_3']
for i in null_values_coloumns:
    value_counts_1 = iitg[i].value_counts(normalize=True)
    num_missing_1 = iitg[i].isna().sum()
    fill_values_1 = np.random.choice(value_counts_1.index, size=num_missing_1, p=value_counts_1.values)
    iitg.loc[iitg[i].isna(), i] = fill_values_1
# i used this approcah bcoz it will not change data pattern bcoz it is a categorical data 
```

```python
# it have large / more values so used median
iitg['Feature_2'] = iitg['Feature_2'].fillna(iitg['Feature_2'].median())
iitg['absences'] = iitg['absences'].fillna(iitg['absences'].median())
```

# level 4

```python
from sklearn.metrics import classification_report, accuracy_score
X= iitg.values[:,0:31]
Y= iitg.values[:,32]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, 
                                                    random_state = 100)
```

## Decision Tree

```python

X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)
X_test_encoded = X_test_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)
clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state = 100,max_depth=3,min_samples_leaf=5)
clf_entropy.fit(X_train_encoded, y_train)
```

```python

y_pred = clf_entropy.predict(X_test_encoded)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
Rf = RandomForestClassifier(n_estimators=100, random_state=100)
Rf.fit(X_train, y_train)
```

```python
y_pred = Rf.predict(X_test_encoded)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
